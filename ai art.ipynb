```python
import torch
import gradio as gr
from diffusers import StableDiffusionXLPipeline, EulerAncestralDiscreteScheduler
from PIL import Image, ImageDraw, ImageFont
from huggingface_hub import login
from deep_translator import GoogleTranslator
from langdetect import detect  # Language detection for better translation

# ✅ Free up memory before starting
torch.cuda.empty_cache()

# ✅ Authenticate with Hugging Face
login(token="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")  # Replace with your actual token

# ✅ Load Stable Diffusion XL
model_id = "stabilityai/stable-diffusion-xl-base-1.0"
pipe = StableDiffusionXLPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    use_safetensors=True,  # Enable safe tensor loading
    cache_dir="./sd_cache",  # Use cache to prevent re-downloading
    resume_download=True,    # Resume if interrupted
)
pipe.scheduler = EulerAncestralDiscreteScheduler.from_config(pipe.scheduler.config)
pipe.to("cuda", torch_dtype=torch.float16, device_map="auto")  # ✅ Optimized GPU allocation

# ✅ Function to translate & shorten text
def translate_text(text):
    detected_lang = detect(text)  # Detect language
    print(f"Detected Language: {detected_lang}")  # Debugging output

    if detected_lang == "en":  # If already English, no need to translate
        return text

    translated_text = GoogleTranslator(source="auto", target="en").translate(text)
    return translated_text[:47] + "..." if len(translated_text) > 50 else translated_text

# ✅ Function to overlay short English text on image
def add_text_to_image(image, text):
    translated_text = translate_text(text)  # Ensure English text

    draw = ImageDraw.Draw(image)
    try:
        font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 50)
    except:
        font = ImageFont.load_default()

    text_position = (50, 50)
    text_color = (255, 255, 255)
    shadow_color = (0, 0, 0)

    draw.text((text_position[0] + 2, text_position[1] + 2), translated_text, font=font, fill=shadow_color)
    draw.text(text_position, translated_text, font=font, fill=text_color)

    return image

# ✅ Define the optimized image generation function
def generate_image(prompt):
    torch.cuda.empty_cache()  # ✅ Free GPU memory before generating
    try:
        translated_prompt = translate_text(prompt)  # Ensure prompt is in English
        print(f"Using translated prompt: {translated_prompt}")  # Debugging output
        image = pipe(translated_prompt, num_inference_steps=30, guidance_scale=7.5).images[0]
        return add_text_to_image(image, translated_prompt)
    except torch.cuda.OutOfMemoryError:
        return "⚠️ GPU memory ran out! Try reducing image size or restarting runtime."

# ✅ Create the Gradio interface
iface = gr.Interface(
    fn=generate_image,
    inputs=gr.Textbox(label="Enter your prompt"),
    outputs=gr.Image(label="Generated Image"),
    title="✨ High-Quality AI Image Generator",
    description="Enter a text prompt to generate an **elegant, fantastic AI image with an English text overlay!**"
)

# ✅ Launch the Gradio app
iface.launch(share=True)
```
